**progress(** x, value, label, min, max, step, color, orientation **)**

```python
def progress(x = "float", value = 0, label = "", min = 0, max = 100, step = 0.1, color = "green", orientation = "horizontal"):
    if x == "int":
        return widgets.IntProgress(
            value = int(value), description = label, min = min, max = max, step = step, style={'bar_color': f"{color}"}, orientation = orientation, )

    if x == "float":
        return widgets.FloatProgress(
            value = float(value), description = label, min = min, max = max, step = step, style={'bar_color': f"{color}"}, orientation = orientation)


# progress

g = progress("float", 75, "", 0, 100, 0.01, "green", orientation = "horizontal")
h = progress("int", 25, "", 0, 100, 0.01, "red", orientation = "horizontal")

display(g, h)
```
